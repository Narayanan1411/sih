```python
import os

dataset_path = "/content/drive/MyDrive/dataset/archive (15)/16000_pcm_speeches"
data = []

for person in os.listdir(dataset_path):
    person_path = os.path.join(dataset_path, person)
    if os.path.isdir(person_path):
        for file in os.listdir(person_path):
            if file.endswith(".wav"):
                file_path = os.path.join(person_path, file)
                data.append((person, file_path))

# Print the extracted data
dataset=[]
for person, file_path in data:
    features=extract_mel_spectrogram(file_path)
    dataset.append([person,features])
```

```python
import librosa.display
import numpy as np
def extract_mel_spectrogram(audio_path, sr=22050, n_mels=128):
    y, sr = librosa.load(audio_path, sr=sr)
    mel_spec = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=n_mels)
    return np.mean(mel_spec, axis=1)
```

```python
from tensorflow import keras
from keras.layers import *
from keras.models import Sequential
```

```python
input_size=(16,16)
```

```python
model=Sequential()
model.add(GRU(32,input_shape=input_size,return_sequences=True))
model.add(Dropout(0.5))
model.add(GRU(64,return_sequences=True))
model.add(Dropout(0.2))
model.add(GRU(64,return_sequences=False))
model.add(Dense(32))
model.add(Dense(6,activation='sigmoid'))
```

```python
model.compile(loss='Categorical_crossentropy',metrics=['accuracy'],optimizer='adam')
```
